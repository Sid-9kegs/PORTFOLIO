# Guess The Number Game

#### 1. Setting up the libraries and explaining the rules of the Game.

```python
import random 
print()
print(f'''let's play a game!!''')
print()
print(f''' RULES \n
1. The computer will pick a random number between 1 and 20. \n
2. Your job is to guess the numbers in the fewest possible tries.\n
3. All the numbers in this game will be whole numbers.\n 
4. Compare Score with friends.\n
5. The player with the lowest score wins''')
print()
```

#### 2. Function definition and iterating through different scenarios.

```python
def rangame(x):
    ran  = random.randint(1,x)
    guess = 0
    count = 0
    while True:
        
        guess = input('Enter your guess: ')
        if len(guess) ==0 or guess == 'done':
            break
          
        try:
            y = int(guess)
        except:
            print()
            print('Please enter only whole numbers')   
            continue 
        count = count +1      
        if y> ran:
            print()
            print('too far ahead, guess again.')
            print()
        elif y< ran:
            print()
            print('too low, pull up!!!')
            print()
        elif y== ran:
            print()
            
            print('Yay you won the game!!!!!')
            print()
            print(f'################# Your score is {count}##################')
            break
```

#### 3. Play the Game.

```python
rangame(20) 
```
