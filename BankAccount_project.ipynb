# Python Bank Account Project 


### Demonstrates classes, inheritance, exceptions, and transactions for a simple banking system. This notebook is structured for readability and recruiter review.

######   Base Class: `BankAccount`


```python
class BankAccount:
    def __init__(self, initial_amount, acct_name):
        self.balance = initial_amount
        self.name = acct_name
        print(f"{self.name} created. Balance: ${self.balance:.2f}")

    def get_balance(self):
        print(f"{self.name} balance: ${self.balance:.2f}")

    def deposit(self, amount):
        self.balance += amount
        print("Deposit complete.")
        self.get_balance()

    def viable_transaction(self, amount):
        """Raise BalanceException if the account does not have enough funds."""
        if self.balance < amount:
            raise BalanceException(
                f"Insufficient funds: account only has ${self.balance:.2f}"
            )

    def withdraw(self, amount):
        try:
            self.viable_transaction(amount)
            self.balance -= amount
            print("Withdrawal complete.")
            self.get_balance()
        except BalanceException as error:
            print(f"Interrupted: {error}")

    def transfer(self, amount, account):
        try:
            print("Transferring...")
            self.viable_transaction(amount)
            self.withdraw(amount)
            account.deposit(amount)
            print("Transfer complete.")
        except BalanceException as error:
            print(f"Transfer interrupted: {error}")
```

###### `InterestRewardsAcct` – Inheriting from `BankAccount`
###### `InterestRewardsAcct` extends `BankAccount` and overrides `deposit` to add a 5% bonus to every deposit.

```python
class InterestRewardsAcct(BankAccount):
    def deposit(self, amount):
        self.balance += amount * 1.05  # 5% bonus
        print("Deposit complete (with 5% bonus).")
        self.get_balance()
```

###### `SavingsAcct` – Fees with Inheritance
###### `SavingsAcct` builds on `InterestRewardsAcct` and introduces a fixed withdrawal fee.  
###### Every withdrawal checks that the balance can cover `amount + fee` before proceeding.

```python
class SavingsAcct(InterestRewardsAcct):
    def __init__(self, initial_amount, acct_name):
        super().__init__(initial_amount, acct_name)
        self.fee = 5

    def withdraw(self, amount):
        try:
            self.viable_transaction(amount + self.fee)
            self.balance -= (amount + self.fee)
            print("Withdrawal complete (fee applied).")
            self.get_balance()
        except BalanceException as error:
            print(f"Interrupted: {error}")
            
```

###### Demo: Creating Accounts and Running Transactions

###### The following examples create several accounts and demonstrate deposits, withdrawals, transfers, and error handling when balances are insufficient.

```python
# Create base accounts
dave = BankAccount(1000, "Dave")
sara = BankAccount(2000, "Sara")

# Basic operations
dave.get_balance()
sara.get_balance()
sara.deposit(500)

# Withdrawals (successful and failing)
dave.withdraw(10)       # Should succeed
dave.withdraw(10000)    # Should raise BalanceException

# Transfer between base accounts
dave.transfer(100, sara)

# InterestRewards account
jim = InterestRewardsAcct(1000, "Jim")
jim.get_balance()
jim.deposit(100)
jim.transfer(100, dave)

# Savings account with fee
blaze = SavingsAcct(1000, "Blaze")
blaze.get_balance()
blaze.deposit(100)
blaze.transfer(1000, sara)  # Likely to fail due to fee
```
